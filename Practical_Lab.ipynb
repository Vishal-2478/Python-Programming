### Given a list of words, return a dictionary where
### the key is the word and the value is the number of times it occurs in the list.


```python
def count_words(l):
    d = {}

    for word in l:
        word = word.lower()
        if word in d:
            d[word] += 1
        else:
            d[word] = 1

    return d

l = ["vishal", "Keshav", "Rohan", "Vishal", "Keshav", "Yash", "Vishal", "yash"]
print("Counted words : ", count_words(l))
```

### Write a Python program to generate list with 'width'-bit gray code.
### Sample Output: ['00', '01', '11', '10'] ['000', '001', '011', '010', '110', '111', '101', '100']

```python
def gray_code(width):

    codes = ["0", "1"]

    for _ in range(2, width + 1):
        ans = []

        for c in codes:
            ans.append("0" + c)

        for c in reversed(codes):
            ans.append("1" + c)

        codes = ans

    return codes

print("Gray code for 2 : ", gray_code(2))
print("Gray code for 3 : ", gray_code(3))
```

### Write a Python program to find the first triangle number to have over n(given) divisors

```python
n = int(input("Enter number of divisors: "))

i = 1

while True:
    tri = i * (i + 1) // 2
    divCount = 0

    d = 1
    while d * d <= tri:
        if tri % d == 0:
            if d * d == tri:
                divCount += 1
            else:
                divCount += 2
        d += 1

    if divCount > n:
        print(f"First triangular number whose divisors greater than {n} is {tri}")
        break

    i += 1
```
